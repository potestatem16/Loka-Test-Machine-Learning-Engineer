# Data preparation and Pre Processing

In the context of the CRISP-DM methodology, the Data Preparation phase for processing AWS documentation files involved several crucial transformations to ensure the data was suitable for the intended analysis and modeling, particularly for training a BERT algorithm. Initially, the raw data, primarily in Markdown format, required cleaning and normalization. This step was essential to remove or convert Markdown and HTML tags, ensuring that the text was in a clean and consistent format. This process not only facilitated easier parsing and analysis of the text but also ensured that the input to the BERT model was free from irrelevant syntactic elements that could hinder its performance.

Further, the data underwent a structuring process where it was divided into logical sections based on headings and subheadings. This segmentation was critical for creating a structured format that could be efficiently indexed and retrieved later. Additionally, metadata tagging was applied, where relevant information like document titles, section headings, and links were identified and tagged. This metadata played a pivotal role in enhancing the information retrieval process, allowing for more precise and contextually relevant responses to user queries. These transformations were aimed at refining the raw data into a format that was not only conducive to training the BERT model effectively but also aligned with the end goal of developing a robust document retrival system.

**transform JSON structures**

While JSON is great for data storage and transmission due to its hierarchical structure, it’s not ideal for text processing models that need a linear, readable format. By transforming JSON into plain text, we enable the model to interpret the information in a way that mirrors natural human language, leading to better context comprehension and more precise content processing.

````python
def json_to_plain_text(json_data):
    # Function to process dictionary objects in JSON
    def process_dict(d, indent=0):
        plain_text = ""
        for key, value in d.items():
            plain_text += " " * indent
            plain_text += f"{key}: "
            if isinstance(value, dict):
                # Recursively process nested dictionaries
                plain_text += "\n" + process_dict(value, indent + 4)
            elif isinstance(value, list):
                # Process lists by recursively processing each element
                plain_text += "[\n" + process_list(value, indent + 4) + " " * indent + "]"
            else:
                # Directly append other types of values
                plain_text += f"{value}\n"
        return plain_text

    # Function to process list objects in JSON
    def process_list(lst, indent=0):
        return "".join(process_dict(item, indent) if isinstance(item, dict) else " " * indent + str(item) + "\n" for item in lst)

    return process_dict(json_data)

def clean_and_validate_json(text):
    # Add brackets if necessary
    if not (text.strip().startswith('{') and text.strip().endswith('}')):
        text = '{' + text.strip() + '}'
    # Clean the JSON text
    text = clean_json_string(text)
    # Check if the text is valid JSON
    if is_valid_json(text):
        return text
    else:
        return None

def clean_json_string(json_str):
    # Remove extra commas at the end of objects or arrays
    json_str = re.sub(r',\s*}', '}', json_str)
    json_str = re.sub(r',\s*\]', ']', json_str)
    return json_str

def is_valid_json(text):
    try:
        json.loads(text)
        return True
    except json.JSONDecodeError:
        return False

def detect_and_transform_json_structures(content):
    json_regex = r'`\n([\s\S]*?)\n`'
    json_matches = re.findall(json_regex, content)

    # Validate and clean JSON matches
    json_matches_valid = [clean_and_validate_json(match) for match in json_matches if clean_and_validate_json(match) is not None]

    for i in range(len(json_matches_valid)):
        try:
            json_data = json.loads(json_matches_valid[i])
            json_text = json_to_plain_text(json_data)

            pattern = re.escape('`\n' + json_matches[i] + '\n`')
            # Replace the original JSON structure with plain text in the content
            content = re.sub(pattern, json_text, content)
        except json.JSONDecodeError:
            # Ignore if it's not valid JSON
            pass

    return content
````

**transform directory structures**

The provided code functions to transform directory structures, often represented in a tree-like format, into a more readable plain text description. This transformation is essential for natural language processing (NLP) models, as these models are designed to process and understand human-readable text rather than structured or symbolic representations like directory trees.

````python
def directory_to_text(directory_structure):
    # Split the directory structure into individual lines
    lines = directory_structure.strip().split('\n')
    text_description = ""

    for line in lines:
        # Remove initial spaces and directory structure characters
        clean_line = line.strip().lstrip("├── └── │")

        if clean_line:  # If the line is not empty
            if line.startswith("├──") or line.startswith("└──"):
                # It's a main directory or file
                if "├──" in line or "└──" in line:
                    # Add a description for a main directory or file
                    text_description += f"\nThe directory '{clean_line}' contains: "
                else:
                    # Add a description for a standalone directory or file
                    text_description += f"\nThe directory '{clean_line}'"
            elif "└──" in line or "├──" in line:
                # It's a subdirectory or file within a directory
                text_description += f"\n - '{clean_line}'"
            else:
                # It's part of the directory or file structure
                text_description += f", '{clean_line}'"

    return text_description

# Function to detect and transform directory structures in the content
def detect_and_transform_directory_structures(content):
    # Regex to find directory structures in the content
    directory_matches = re.findall(r'(`{1,3}\n?[\s\S]*?\n?`{1,3})', content)
    for directory_match in directory_matches:
        try:
          # Remove backticks to isolate the directory structure
          directory_structure = directory_match.strip('`')
          # Transform the directory structure to text
          directory_text = directory_to_text(directory_structure)
          # Replace the original directory structure with the transformed text
          content = content.replace(directory_match, directory_text)
        except:
          # Ignore if there's an error in processing
          pass
    return content
````

Next, the code below consists of two functions designed to process and transform Markdown content, particularly useful when preparing such data for natural language processing (NLP) tasks. The first function, transform_markdown, performs a series of transformations on the Markdown content to make it more suitable for NLP models, while the second function, process_markdown_folder, applies these transformations to all Markdown files in a specified folder and compiles the results into a pandas DataFrame.

## transform_markdown Function:
1. Transform Titles and Subsections:
The function uses regular expressions to identify Markdown headings ```(#, ##, ###)``` and converts them into a more readable format. For example, top-level headings (#) are prefixed with "Documentation Title:", second-level headings (##) with "Subsection:", and third-level headings (###) with "Subsubsection:". This transformation makes the structure of the document clearer and more accessible for text processing.

2. Remove Special Markdown Characters:
Markdown formatting like bold ```(**text**)```, italic ```(*text*)```, and inline code ```( text )``` is removed, leaving only the plain text. This step is crucial because NLP models are typically designed to work with plain text and might get confused by Markdown syntax.

3. Transform Lists:
Bullet points in lists, marked by - or *, are replaced with the phrase "List Element:". This change standardizes list items, making them more recognizable in the transformed text.

4. Retain Only Web Links as "Interest Links":
Hyperlinks in Markdown, formatted as ```[link text](URL)```, are transformed to keep only the URL with the prefix "Interest link:". This approach simplifies the representation of hyperlinks, focusing on the URL, which is often the most relevant part for NLP tasks.

5. Remove HTML Metadata and Comments:
Any HTML anchors ```(<a name="..."></a>)``` and comments ```(<!-- comment -->)``` within the Markdown content are removed. These elements are typically not relevant for text analysis and can be safely discarded

````python
# Main function to transform Markdown content
def transform_markdown(content):
    # Transform headers and subsections
    # Convert Markdown headers into a more descriptive format
    content = re.sub(r'^#{1} (.*)', r'Documentation Title: \1', content, flags=re.MULTILINE)
    content = re.sub(r'^#{2} (.*)', r'Subsection: \1', content, flags=re.MULTILINE)
    content = re.sub(r'^#{3} (.*)', r'Subsubsection: \1', content, flags=re.MULTILINE)

    # Remove special Markdown characters
    # This includes bold, italic, and inline code formatting
    content = re.sub(r'\*\*(.*)\*\*', r'\1', content)
    content = re.sub(r'\*(.*)\*', r'\1', content)
    content = re.sub(r'`(.*)`', r'\1', content)

    # Transform lists
    # Convert Markdown list items into a more readable format
    content = re.sub(r'\n\s*[-\*] (.*)', r'\nList Element: \1', content)

    # Keep only web links as "Interest links"
    # Extract and reformat hyperlinks
    content = re.sub(r'\[(.*?)\]\((http[s]?://.*?)\)', r'Interest link: \2', content)

    # Remove HTML metadata and comments
    # Clean up HTML artifacts that may be present in Markdown
    content = re.sub(r'<a name=".*?"></a>', '', content)
    content = re.sub(r'<!--.*?-->', '', content)

    return content

# Function to process a folder of Markdown files
def process_markdown_folder(folder_path):
    data = []
    for filename in os.listdir(folder_path):
        print(filename)
        if filename.endswith('.md'):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                # Apply the transformation to the Markdown content
                transformed_content = transform_markdown(content)
                # Append the original and transformed content along with the filename
                data.append({'filename': filename, 'original content': content, 'transformed content': transformed_content})
    # Convert the list of data into a pandas DataFrame
    return pd.DataFrame(data)
````

The next code consists of two functions designed for handling text data and performing semantic search using embeddings generated by a Sentence-BERT model.

**Function search:**
* This function performs a semantic search to find the most similar documents to a given query.
* It takes a query string and an optional top_k parameter (defaulting to 2), which specifies the number of top results to return.
* The function generates an embedding for the query and computes cosine similarity scores between this query embedding and the embeddings of all documents.
* It then selects the top k documents with the highest similarity scores.
* For each of these top results, the function retrieves the corresponding filename and content from the DataFrame df, truncates the content to 1800 tokens for readability, and appends this information to the results text.
* The final results text, containing the filenames and truncated contents of the top documents, is returned.

````python
from transformers import PreTrainedTokenizerFast

# BERT tokenizer
tokenizer = PreTrainedTokenizerFast.from_pretrained("bert-base-uncased")

def truncate_to_token_limit(text, token_limit=1800):
    # Tokenize the input text
    tokens = tokenizer.tokenize(text)
    # Truncate the tokens to the specified limit
    truncated_tokens = tokens[:token_limit]
    # Convert the truncated tokens back to a string
    return tokenizer.convert_tokens_to_string(truncated_tokens)


def search(query, top_k=2):
    # Generate an embedding for the query
    query_embedding = generate_embeddings([query])[0]
    # Compute cosine similarity scores between the query and all document embeddings
    cos_scores = util.pytorch_cos_sim(query_embedding, document_embeddings)[0]
    # Select the top k results
    top_results = torch.topk(cos_scores, k=top_k)

    results_text = f"\nTop {top_k} results:\n"

    for score, idx in zip(top_results[0], top_results[1]):
        idx = idx.item()  # Convert tensor to integer
        # Append the filename of the document
        results_text += df.iloc[idx]['filename'] + "\n"
        # Retrieve and truncate the content of the document
        document_content = df.iloc[idx]['transformed_JSON_dir content']
        truncated_content = truncate_to_token_limit(document_content, 1800)
        # Append the truncated content to the results
        results_text += truncated_content + "\n\n"

    return results_text
````

````python
%%writefile app.py

import streamlit as st
import replicate
import os
import torch
import pandas as pd
import numpy as np
import transformers
from sentence_transformers import SentenceTransformer, util
from PIL import Image
os.getcwd()

os.chdir(r"G:\My Drive\BERT Modelo de Lenguaje\Loka")


document_embeddings = np.load('document_embeddings.npy')

df = pd.read_csv("AWS sagemaker documentation.csv")


from transformers import PreTrainedTokenizerFast

tokenizer = PreTrainedTokenizerFast.from_pretrained("bert-base-uncased")

def truncate_to_token_limit(text, token_limit=1800):
    tokens = tokenizer.tokenize(text)
    truncated_tokens = tokens[:token_limit]
    return tokenizer.convert_tokens_to_string(truncated_tokens)



def search(query, top_k=2):
    query_embedding = generate_embeddings([query])[0]
    cos_scores = util.pytorch_cos_sim(query_embedding, document_embeddings)[0]
    top_results = torch.topk(cos_scores, k=top_k)

    results_text = f"\nTop {top_k} results, but include also results to the best of your knowledge, consistent with the following documentation. Provide the documentation titles of interest:\n"

    filenames = []

    for score, idx in zip(top_results[0], top_results[1]):
        idx = idx.item()
        filename = df.iloc[idx]['filename']
        filenames.append(filename)
        results_text += filename + "\n"
        document_content = df.iloc[idx]['transformed_JSON_dir content']
        truncated_content = truncate_to_token_limit(document_content, 1800)
        results_text += truncated_content + "\n\n"

    return results_text, filenames


# Sentence Transformer
model = SentenceTransformer('all-mpnet-base-v2')#.to(device)

def generate_embeddings(texts):

    embeddings = model.encode(texts, convert_to_tensor=True)
    return embeddings.to(device)




os.environ['REPLICATE_API_TOKEN'] = ''


api = replicate.Client(api_token=os.environ["REPLICATE_API_TOKEN"])


import streamlit as st
import replicate
import os
from PIL import Image

# Set up the page configuration and title
st.set_page_config(page_title="🦙💬 Llama2/BERT AWS Sagemaker Assistant Assistant")

# Sidebar for Replicate credentials and model parameters
with st.sidebar:
    # Display an image in the sidebar
    image = Image.open('AWS.png')
    st.image(image, width=300)

    # Display a title in the sidebar
    st.title('🦙💬 Llama2/BERT AWS Sagemaker Assistant Assistant \n Loka Test from Manuel Diaz')

    # Check if API key is provided and validate it
    if 'REPLICATE_API_TOKEN' in st.secrets:
        st.success('API key already provided!', icon='✅')
        replicate_api = st.secrets['REPLICATE_API_TOKEN']
    else:
        replicate_api = st.text_input('Enter Replicate API token:', type='password')
        if not (replicate_api.startswith('r8_') and len(replicate_api)==40):
            st.warning('Please enter your credentials!', icon='⚠️')
        else:
            st.success('Proceed to entering your prompt message!', icon='👉')
    os.environ['REPLICATE_API_TOKEN'] = replicate_api

    # Sidebar options for selecting model and setting parameters
    selected_model = st.selectbox('Choose a Llama2 model', ['Llama2-7B', 'Llama2-13B'])
    temperature = st.slider('temperature', min_value=0.01, max_value=5.0, value=0.1, step=0.01)
    top_p = st.slider('top_p', min_value=0.01, max_value=1.0, value=0.9, step=0.01)
    max_length = st.slider('max_length', min_value=32, max_value=128, value=120, step=8)

# Initialize chat history
if "messages" not in st.session_state.keys():
    st.session_state.messages = [{"role": "assistant", "content": "How may I assist you today?"}]

# Display chat messages and provide an option to clear chat history
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.write(message["content"])
def clear_chat_history():
    st.session_state.messages = [{"role": "assistant", "content": "How may I assist you today?"}]
st.sidebar.button('Clear Chat History', on_click=clear_chat_history)

# Function to generate responses using Llama2 model
def generate_llama2_response(prompt_input):
    # Prepare dialogue string and add context search to the prompt
    search_result, filenames = search(prompt_input)  # Get filenames from the search function
    prompt_input = "Question: " + prompt_input + "\n\n" + search_result

    # Run the Llama2 model via Replicate API
    output = api.run(
        llm,
        input={
            "debug": False,
            "top_k": 50,
            "top_p": top_p,
            "prompt": prompt_input,
            "temperature": temperature,
            "system_prompt": "You are an assistant focused on helping programmers about AWS Sagemaker Documentation, and all related questions",
            "max_new_tokens": 500,
            "min_new_tokens": -1
            }
        )

    return output, filenames

# User input prompt
if prompt := st.chat_input(disabled=not replicate_api):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.write(prompt)

# Generate a new response if the last message is not from the assistant
if st.session_state.messages[-1]["role"] != "assistant":
    with st.chat_message("assistant"):
        with st.spinner("Thinking..."):
            response, filenames = generate_llama2_response(prompt)  # Get filenames from the generate_llama2_response function
            placeholder = st.empty()
            full_response = ''
            for item in response:
                full_response += item
                placeholder.markdown(full_response)
            # Add filenames as "sources" at the end of the response
            full_response += "\n" + "\nSources:\n" + "\n".join("* " + filename for filename in filenames)
            placeholder.markdown(full_response)
    message = {"role": "assistant", "content": full_response}
    st.session_state.messages.append(message)


````
